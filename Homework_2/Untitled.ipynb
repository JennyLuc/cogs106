```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
daily_sales = pd.read_csv("daily_sales.csv")
daily_sales = daily_sales.rename(columns={'Unnamed: 0':'Day_of_Week'})
daily_sales.head()
```

```python
monthly_sales = pd.read_csv("monthly_sales.csv")
monthly_sales.head()
```

```python
def day (date):
    return date.split('/')[1]
```

```python
daily_ne = pd.DataFrame(
    data = {'Day' : daily_sales['Day_of_Week'],
            'Date': daily_sales['Day'], 
            'hm' : daily_sales['HM-NE'], 
            'cf' : daily_sales['CF-NE'],
            'ff' : daily_sales['FF-NE']})
daily_ne['Date'] = daily_ne['Date'].apply(day)
```

```python
daily_ne.head()
```

```python
reg_line = daily_ne.plot(x = 'Date', y = ['hm', 'cf', 'ff'], kind = 'line', 
                        title = 'line graph of food sold daily in the North East', figsize = (15,8), grid = True)
reg_line.set_ylabel('USD')
```

```python
hm_daily = []
ff_daily = []
cf_daily = []
for i in range(len(daily_sales)):
    hm = 0
    ff = 0
    cf = 0
    for col in daily_sales.loc[[i]].columns:
        if 'HM' in col:
            hm += int(daily_sales.loc[[i]][col][i])
        if 'FF' in col:
            ff += daily_sales.loc[[i]][col][i]
        if 'CF' in col:
            cf += daily_sales.loc[[i]][col][i]    
    if hm > 0:
        hm_daily.append(hm)
    if ff > 0:
        ff_daily.append(ff)
    if cf > 0:
        cf_daily.append(cf)

```

```python
month_nation = pd.DataFrame(data = {'Date':daily_se['Date'], 'hamburger' : hm_daily, 'fish_fillet': ff_daily, 'chicken_fillet': cf_daily})
month_nation.head()
```

```python
df_month_nation = month_nation.plot(kind = 'line', x = 'Date', y =['hamburger',  'chicken_fillet', 'fish_fillet'], grid = True,
                 title = 'Daily Line Graph of the Nation', figsize = (15,8))
df_month_nation.set_ylabel('USD')
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(daily_se['Date'], hm_daily, label='hamburger')
ax.plot(daily_se['Date'], ff_daily, label='fish fillet')
ax.plot(daily_se['Date'], cf_daily, label='chicken fillet')
ax.legend(loc='upper center',bbox_to_anchor=(1.25, 0.8), ncol=1)
plt.title('daily line graph of the nation')
```

```python
df = monthly_sales.copy()
df = df.rename(columns = {'Month, Year': 'Date'})
df['Date'] = pd.to_datetime(df['Date'], format = '%b-%y')
df = df.sort_values(by=['Date'])
df['Date'] = df['Date'].dt.strftime('%b-%y')
```

```python
hm_month_df = df.groupby(monthly_sales['Month, Year']).aggregate(sum)
hm_month_df = hm_month_df.reindex(df['Date'])
hm_month_df = hm_month_df[['HM-NE', 'HM-NW', 'HM-SW', 'HM-SE', 'HM-C']]
hm_month_df_line = hm_month_df.plot(kind = 'line', figsize = (18,10),
                                   title = 'line graph of hamburgers in months', grid = True)
hm_month_df_line.set_ylabel('USD')
```

```python
cf_month_df = df[['CF-NE', 'CF-NW', 'CF-SW', 'CF-SE', 'CF-C']]
cf_month_df = cf_month_df.groupby(monthly_sales['Month, Year']).aggregate(sum)
cf_month_df = cf_month_df.reindex(df['Date'])
cf_month_df_line = cf_month_df.plot(kind = 'line', figsize = (18,10),
                                   title = 'line graph of chicken fillets sold in months', grid = True)
cf_month_df_line.set_ylabel('USD')
```

```python
ff_month_df = df[['FF-NE', 'FF-NW', 'FF-SW', 'FF-SE', 'FF-C']]
ff_month_df = ff_month_df.groupby(monthly_sales['Month, Year']).aggregate(sum)
ff_month_df = ff_month_df.reindex(df['Date'])
ff_month_df_line = ff_month_df.plot(kind = 'line', figsize = (18,10),
                                   title = 'line graph of fish fillets sold in months', grid= True)
ff_month_df_line.set_ylabel('USD')
```
