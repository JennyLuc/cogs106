```python
heat_df = pd.read_csv ('dsc106.csv')
sport_2008 = heat_df[(heat_df['year']==2008)]
sport_1948 = heat_df[(heat_df['year']==1948)]
len(sport_1948.sport.unique())
```

```python
sport_2008.sport.unique()
```

```python
sport = pd.merge(sport_2008, sport_1948, how = 'inner', on = 'sport')
sport_list = list(sport.sport.unique())
year_list = list(heat_df.year.unique())
```

```python
heat_df.head()
```

```python
new_heat_df = heat_df[heat_df.sport.isin(sport_list)][['year', 'sport', 'sex']]
female = new_heat_df[new_heat_df.sex == "F"].groupby(['sport','year'], as_index = False).count()
total = new_heat_df.groupby(['sport','year'], as_index = False).count()
female.head()
```

```python
total.head()
```

```python
test = pd.merge(total, female, on = ['sport', 'year'], how = 'inner')
test['proportion'] = test['sex_y']/test['sex_x']
test = test.drop(['sex_x', 'sex_y'], axis = 1)
test.head()
```

```python
tester = pd.DataFrame(index = (test.year.unique()), columns = list(test.sport.unique()))
```

```python
for y,s,p in zip(test.year, test.sport, test.proportion):
    tester.loc[[y],s] = round(p,2)
```

```python
tester.sort_index(inplace=True, ascending = False)
tester = tester.fillna(0)
tester
```

```python
import seaborn as sns; sns.set()
plt.figure(figsize=(10, 10), dpi = 200)

cmap = sns.cubehelix_palette(dark=0, light=1, as_cmap=True)
sns_plot = sns.heatmap(tester, cmap = cmap, square = True)
plt.title('Heatmap of Female Participation Proportions')
plt.xlabel('Sport')
plt.ylabel('Year')
fig = sns_plot.get_figure()
fig.savefig("heatmap.png")
```
